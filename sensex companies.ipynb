```python
import pandas as pd
```

```python
sp_url = 'https://en.wikipedia.org/wiki/List_of_BSE_SENSEX_companies'
sp = pd.read_html(sp_url, header=0)[0] # returns a list for each table
sp.info()
```

```python
csv_filename = 'sensex_data.csv'
sp.to_csv(csv_filename, index=False)
print(f'Data has been saved to {csv_filename}')
```
